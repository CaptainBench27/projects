```python
import pandas as pd
url = "https://raw.githubusercontent.com/lilaceri/porfolio/main/datasets/Netflix%20viewing%20grannyyp.csv"
netflix = pd.read_csv(url)

df = netflix.copy()

series = df[(df["Title"].str.contains(": Season")) | (df["Title"].str.contains(": Series")) | (df["Title"].str.contains(": Episode"))| (df["Title"].str.contains(": Volume"))]
film = df[(~df["Title"].str.contains(": Season")) & (~df["Title"].str.contains(": Series")) & (~df["Title"].str.contains(": Episode")) & (~df["Title"].str.contains(": Volume"))]
series['Episode'] = series["Title"].str.split(":").str.get(2)
series["Name"] = series['Title'].str.split(":").str.get(0)
series['Season'] = series['Title'].str.split(":").str.get(1)


# np_series = series['Episode'].to_numpy()
# for i in np_series:
#   print(i)

```
